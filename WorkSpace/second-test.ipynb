## evaluation on the validationset

```python
evaluation = model.val(
    data = '/kaggle/working/Road-Lines-Segmentation-10/data.yaml',
    task = 'segment',
    imgsz = 640,
    iou = 0.6,
    batch = 32,
    device = 0,
    verbose = True,
    seed = 93,
    plots = True,
    name = 'Model-segment-evaluation',
    save_json = True
)
```

## predictions on the testset

```python
predictions = model.predict(
    data = '/kaggle/working/Road-Lines-Segmentation-10/data.yaml',
    source = '/kaggle/working/Road-Lines-Segmentation-10/test/images',
    task = 'segment',
    imgsz = 640,
    #device = 0,
    verbose = True,
    seed = 93,
    plots = True,
    name = 'Model-segment-predictions',
    save_json = True,
)
```
